```python
def generateObjective1():
    c = np.zeros(678, dtype=np.int64)

    assign(c, 1, 297, -np.tile(profit, 9))
    
    return c

def generateObjective2():
    c = np.zeros(678, dtype=np.int64)

    assign(c, 645, 645, -162)
    
    return c

def generateObjective3():
    c = np.zeros(678, dtype=np.int64)

    assign(c, 265, 297, -2 * area)
    
    return c

def generateObjective4():
    c = np.zeros(678, dtype=np.int64)

    assign(c, 646, 678, 2 * area)
    
    return c

def generateObjective5():
    c = np.zeros(678, dtype=np.int64)

    assign(c, 265, 297, cost)
    
    return c

def generateObjective6():
    c = np.zeros(678, dtype=np.int64)

    assign(c, 628, 636, -1)
    
    return c
```

```python
c = [
    generateObjective1(),
    generateObjective2(),
    generateObjective3(),
    generateObjective4(),
    generateObjective5(),
    generateObjective6(),
]

c = np.array(c, dtype=np.int64)

print(c.shape)
print(c)

c = c.sum(axis=0)
print(c.shape)
print(c)
```

```python
X_bounds = []
X_bounds.extend([(0,1) for _ in range(33 * 9 * 2 + 33)]) # X, Y, z
X_bounds.extend([(0,100) for _ in range(9)]) # ψ
X_bounds.extend([(0,500) for _ in range(8)]) # ξ
X_bounds.extend([(0,1)]) # λ
X_bounds.extend([(0,1) for _ in range(33)]) # ω
    
print(f'{c.shape = }')
print(f'{A.shape = }')
print(f'{b.shape = }')

start = timer()
res = linprog(c, A_ub=A, b_ub=b, bounds=X_bounds, integrality=1, options={
    'presolve': False, # disable as it can cause errors
    'disp': True,
    'primal_feasibility_tolerance': 1e-10, # use the minimum tolerance possible for both primal and dual
    'dual_feasibility_tolerance': 1e-10
}, method='highs')
end = timer()
print(res)
print(f'Solver took {end - start}s')
```
